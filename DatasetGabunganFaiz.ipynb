```python
import pandas as pd
```

```python
from google.colab import drive
drive.mount('/content/drive')
folder_path = '/content/drive/My Drive/Dataset/'
```

```python
faiz = pd.read_csv(folder_path + 'Recon Ping Sweep.csv')
```

```python
faiz
```

```python
faiz2 = pd.read_csv(folder_path + 'Recon Vulnerability Scan.csv')
```

```python
faiz2
```

```python
faizgabung =pd.concat([faiz,faiz2],ignore_index=True)
```

```python
faizgabung.columns.values
```

```python
x = faizgabung.iloc[:,7: 76]
```

```python
x
```

```python
y = faizgabung.iloc[:,83: 84]
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y, test_size = 0.2, random_state = 42)
```

```python
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
```

```python
faiz = DecisionTreeClassifier(criterion='entropy', splitter = 'random')
faiz.fit(x_train,y_train)
```

```python
y_pred = faiz.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test,y_pred)
```

```python
accuracy_percent = round(accuracy * 100, 2)
print(f"Akurasi: {accuracy_percent}%")
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
fig = plt.figure(figsize = (10, 7))
tree.plot_tree(faiz, feature_names = x.columns.values, class_names = np.array([ 'Recon Ping Sweep','Recon Vulnerability Scan']), filled = True)
plt.show()
```

```python
import seaborn as lol
from sklearn import metrics
label = np.array([ 'Recon Ping Sweep','Recon Vulnerability Scan'])
```

```python
import matplotlib.pyplot as plt
conf_matrix = metrics.confusion_matrix(y_test, y_pred)
plt.figure(figsize=(10, 10))
lol.heatmap(conf_matrix, annot=True, xticklabels=label, yticklabels=label)
plt.xlabel('Prediksi')
plt.ylabel('Fakta')
plt.show()
```
